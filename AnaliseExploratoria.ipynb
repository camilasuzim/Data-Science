```python
#Criando DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Verificando os tipo de Dados
df.dtypes
```

```python
#Qual é a receita total?
df["Valor Venda"].sum()
```

```python
#Qual o custo total?
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
df = df.drop(columns = ["custo"])
```

```python
round(df["Custo"].sum(),2)
```

```python
#Qual é o lucro total?
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head()
```

```python
round(df["Lucro"].sum(), 2)
```

```python
#Criando uma coluna para de tempo de envio
df["Tempo Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

```python
#Média de tempo de envio do produto para cada marca
df["Tempo Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.groupby("Marca")["Tempo Envio"].mean()
```

```python
#Verificando se há dados faltante
df.isnull().sum()
```

```python
#Lucro por ano
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = False)
```

```python
#Gráfico do total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = True).plot.barh(title = "Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto")
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title = "Lucro Anual")
plt.xlabel("Ano")
plt.ylabel("Receita")
```

```python
#Vendas somente de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df["Data Venda"].dt.month)["Lucro"].sum().plot(title = "Lucro Mensal")
plt.xlabel("Mês")
plt.ylabel("Receita")
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title = "Lucro por Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal");
```

```python
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title = "Lucro por Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal");
```
